## 2. Mock (Simülasyon) Veriseti Oluşturma

```python
import os
import json
from PIL import Image

os.makedirs("data/vqa-rad/images", exist_ok=True)
dummy_image = Image.new("RGB", (224, 224), color="gray")
dummy_image.save("data/vqa-rad/images/test_img.jpg")

mock_data = [
    {"image_name": "test_img.jpg", "question": "Are the lungs clear?", "answer": "yes", "answer_type": "closed", "split": "test"},
    {"image_name": "test_img.jpg", "question": "Is there any opacity?", "answer": "no", "answer_type": "closed", "split": "test"},
    {"image_name": "test_img.jpg", "question": "What is the organ?", "answer": "lungs", "answer_type": "open", "split": "test"},
    {"image_name": "test_img.jpg", "question": "Is the heart size normal?", "answer": "yes", "answer_type": "closed", "split": "test"},
    {"image_name": "test_img.jpg", "question": "Where is the abnormality?", "answer": "right lower lobe", "answer_type": "open", "split": "test"}
]
with open("data/vqa-rad/test.json", "w") as f:
    json.dump(mock_data, f)
print("Mock VQA-RAD dataset oluşturuldu!")
```
